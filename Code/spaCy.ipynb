```python
liste_X = []
for ent in doc_X.ents:
    liste_X.append([ent.text, ent.label_])

liste_y = []
for ent in doc_y.ents:
    liste_y.append([ent.text, ent.label_])
    
print(liste_X)
print("\n####################\n")
print(liste_y)
```

```python
dic_X = {}
for ent in doc_X.ents:
    if ent.label_ == "LOC":
        if ent.text in dic_X:
            dic_X[ent.text] += 1
        else:
            dic_X[ent.text] = 1

dic_y = {}
for ent in doc_y.ents:
    if ent.label_ == "LOC":
        if ent.text in dic_y:
            dic_y[ent.text] += 1
        else:
            dic_y[ent.text] = 1

print(dic_X)
print("\n####################\n")
print(dic_y)
```

```python
bn = 0 #bons résultats
cpt = 0 #compteur pour tester sur un échantillon

for i,token in enumerate(liste_X): #on parcourt les mots de x
    
    liste_mots_sim = [] #liste des mots similaires
    similarite = 0 #taux de similarité entre le mot x et le mot y
    mot_x = token[0]
    #print(i,"|",mot_x)
    
    #On cherche les mots 5% en arrière et 5% en avant dans le texte
    border_left = int(i-(len(liste_y)/5))
    if border_left < 0:
        border_left = 0
    border_right = int(i+(len(liste_y)/5))
    if border_right >= len(liste_y):
        border_right = len(liste_y)-1
    
    for mot_y in liste_y[border_left:border_right]: #on parcourt les mots du y
        for n,lettre in enumerate(mot_y[0]): 
            if n >= len(mot_x):
                similarite -= 1
            elif lettre == mot_x[n]: #si même lettre
                similarite += 1
        
        if similarite/len(mot_x) > 0.7: #si y est similaire à x d'au moins 70% on l'ajoute
            liste_mots_sim.append([similarite/len(mot_x),mot_y])
        similarite = 0
    
    liste_mots_sim = sorted(liste_mots_sim, reverse=True)
    print(mot_x,"|",liste_mots_sim,"\n")
    if len(liste_mots_sim) >= 1:
        bn += 1
        #print("%i résultat(s) similaire(s). Le mot le plus proche est: %s\n" %(len(liste_mots_sim),liste_mots_sim[0]))
        #print("%s|%s"%(mot_x,liste_mots_sim[0][1][0]))
    else:
        #print("Aucun résultat similaire aux alentours\n")
        print("%s|######None######"%mot_x)
    
    #cpt+=1
    if cpt==10:
        break
        
print("\n")
print("%i bons résultats sur %i"%(bn,len(liste_X)))
```

```python
import re
import glob
import spacy
import json
import sklearn
import os
import shutil
from sklearn.neighbors import DistanceMetric
from sklearn.feature_extraction.text import CountVectorizer
# from sklearn.cluster import AgglomerativeClustering
# from sklearn.metrics import pairwise_distances
# from sklearn import metrics
# import matplotlib.pyplot as plt
# import numpy as np

def lire_csv(chemin):
    with open(chemin, "r") as f:
        dic = {}
        head = f.readline().split(";")
        
        for label in head:
            dic[label] = []
            
        for ligne in (f.readlines()):
            entity, n = ligne.split(";")
            entity = re.sub("[ \"]","",entity)
            n = int(n[:-1])
            
            for label in head:
                for n in range(n):
                    dic[label].append(entity)
            
    return dic

def liste_resultats(texte, nlp):
    doc = nlp(texte)
    list_resultats =[]
    for ent in doc.ents:
        if ent.label_=="LOC":
            list_resultats.append(ent.text)
    return (list_resultats)

def get_distances(texte1, texte2, N=1, liste_name =["jaccard", "braycurtis","dice", "cosinus"] ):
    dico = {}
    for metric_name in liste_name :
        dico[metric_name] = []
        liste_resultat_dist2 = []
        #for n_max in range(1, N+1):###range([min, default = 0], max, [step, default = 1]) 
        V = CountVectorizer(analyzer='word') 
        X = V.fit_transform([texte1, texte2]).toarray()
        if metric_name!= "cosinus" :  
            dist = DistanceMetric.get_metric(metric_name)     
            distance_tab1=dist.pairwise(X)
            liste_resultat_dist2.append(distance_tab1[0][1])
        else: 
            distance_tab1=sklearn.metrics.pairwise.cosine_distances(X) 
            liste_resultat_dist2.append(distance_tab1[0][1])
        dico[metric_name] = liste_resultat_dist2
    return dico
```
